```python
from collections import Counter
area_count = Counter(area_list)
print(area_count.items())
```

```python
def area_item_gen(area_count, type):
    for item in area_count.items():
        d = {
            'name':item[0],
            'data':[item[1]],
            'type':type
        }
        yield d
```

```python
series = [data for data in area_item_gen(area_count, 'column')]
```

```python
charts.plot(series, show='inline', options=dict(title=dict(text='七日内北京城区二手物品发帖量')))
```

```python
for i in item_info.find():
    date = i['publish_time']
    date_with_year = "2016." + date.replace('-','.')
    item_info.update({'_id':i['_id']}, {'$set':{'publish_time':date_with_year}})
    
```

```python
for i in item_info.find().limit(10):
    print(i)
```

```python
def data_gen(dates, area, limit = 3):
    pipeline = [
        {'$match':{'$and':[{'publish_time':{'$in':dates}},{'area':area}]}},
        {'$group':{'_id':'$type','counts':{'$sum':1}}},
        {'$sort':{'counts':-1}},
        {'$limit':limit}
    ]
    for i in item_info.aggregate(pipeline):
        d = {
            'name' : i['_id'],
            'data' : [i['counts']],
            'type' : 'column'
        }
        yield d
        
```

```python
dates = ['2016.06.02', '2016.06.07']
area = "昌平"
series = [i for i in data_gen(dates,area,5)]
print(series)
options = {
    'chart'   : {'zoomType':'xy'},
    'title'   : {'text': '昌平区发帖数量最大的类目'},
    'subtitle': {'text': '数据图表'},
    'yAxis'   : {'title': {'text': '数量'}}
    }

charts.plot(series,options=options,show='inline')
```
